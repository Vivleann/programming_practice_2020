```python
import numpy as np
import sys

def matrixSort(matrix, column):
    order = (-matrix)[:, column].argsort()

    newMatrix = np.zeros((matrix.shape[0], matrix.shape[1]), dtype=int)
    for i in range(matrix.shape[1]):
        newMatrix[i] = matrix[order[i]]

    return newMatrix

column = int(input())
matrix = np.loadtxt(sys.stdin)
print(matrixSort(matrix, column))
```
